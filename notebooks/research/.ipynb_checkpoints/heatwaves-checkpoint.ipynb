```python

# #Dictionary with years for warming levels
# years_warming_level = {
#     '1': {
#         'rcp26': {'gfdl-esm2m': [2006, 2023]},
#     },
    
# }

# Function definitions for load_temperature_data, calculate_percentile, identify_heatwaves remain the same
from scipy.stats import norm

import xarray as xr
import os

def load_temperature_data(file_path):
    """
    Load temperature data from a netCDF4 file, subset for Vietnam.
    """
    # Define the bounding box for Vietnam
    lat_bounds = slice(30, 8)  # Vietnam's approximate latitude bounds
    lon_bounds = slice(100, 110)  # Vietnam's approximate longitude bounds
    
    # Load the data and select the region of interest
    temperature = xr.open_dataset(file_path)['tas']  # Assuming 'tas' is the temperature variable
    temperature = temperature.sel(lat=lat_bounds, lon=lon_bounds)
    temperature = temperature.sel(time=temperature.time.dt.month.isin([5, 6, 7, 8, 9, 10]))
    
    return temperature


def calculate_percentile(temperature_data, percentile):
    """
    Calculate the specified percentile for the temperature data for each day of the year.
    """

    return temperature_data.quantile(percentile/100, dim="time")



# Base path for the files
base_path = '/nfs/n2o/wcr/szelie/isimip2b'


baseline_99_percentile = {}
# Load baseline temperature data and calculate the 99th percentile
for model in years_warming_level['1']['rcp26']:
    if model == 'hadgem2-es':
        model_str='HadGEM2-ES'
    else:
        model_str=model.upper()
    out_file = f'{base_path}/pct_{model_str}_historical_r1i1p1_EWEMBI_19860101-20051231_vnm.nc4'
    if os.path.exists(out_file):
        continue
    baseline_files = [
        f'{base_path}/tas_day_{model_str}_historical_r1i1p1_EWEMBI_19810101-19901231.nc4',
        f'{base_path}/tas_day_{model_str}_historical_r1i1p1_EWEMBI_19910101-20001231.nc4',
        f'{base_path}/tas_day_{model_str}_historical_r1i1p1_EWEMBI_20010101-20051231.nc4'
    ]
    baseline_data = xr.concat([load_temperature_data(file) for file in baseline_files], dim='time')
    baseline_data = baseline_data.sel(time=slice("-".join(["1986","01","01"]), "-".join(["2005","12","31"])))
    

    baseline_99_percentile[model] = calculate_percentile(baseline_data, 99)
    #baseline_99_percentile[model].to_netcdf(out_file)

```

```python
def round_year(year):
    if year < 2010:
        return 2006
    elif year % 10 == 0 and year > 2010:
        return (year - 10)+1
    else:
        return (year // 10) * 10 +1

def get_relevant_files(years_range, scenario, model, base_path):
    start_year, end_year = years_range
    start_year_rounded = round_year(start_year)
    end_year_rounded = round_year(end_year)
    middle_year_rounded = round_year(start_year+10)
    unique_files_starting_year=np.unique([start_year_rounded,end_year_rounded,middle_year_rounded])
    files = []
    for year in unique_files_starting_year:  # Adjust step based on your file coverage
        if year == 2006:
            end_year_file = 2010
        else:
            end_year_file = year+9
        file_base = f"{base_path}/tas_day_{model}_{scenario}_r1i1p1_EWEMBI_{str(year)}0101-{end_year_file}1231"
        file_path = f"{file_base}.nc4"
        file_path_1 = f"{file_base}.nc4.1"
        if os.path.exists(file_path):
            files.append(file_path)
        # if os.path.exists(file_path_1):
        #     files.append(file_path_1)   
    return files
```

```python
# Initialize a dictionary to store heatwave data
heatwave_data_dict = {}

# Process data for each warming level
for warming_level in ['1','2']:
    
    heatwave_data_dict[warming_level] = {}

    for scenario in years_warming_level[warming_level]:
        heatwave_data_dict[warming_level][scenario] = {}
        
        for model, years in years_warming_level[warming_level][scenario].items():
            print(model)
            if model == 'hadgem2-es':
                model_upper='HadGEM2-ES'
            else:
                model_upper=model.upper()
            if not years:
                continue
            out_file = f'{base_path}/hw_day_{model}_{scenario}_r1i1p1_EWEMBI_{str(years[0])}0101-{years[1]}1231_vnm.nc4'
            # if os.path.exists(out_file):
            #     continue
            # Get relevant files for the specified years
            relevant_files = get_relevant_files(years, scenario, model_upper, base_path)
            temp_data_list = []
            for file in relevant_files:
                temp_data_list.append(load_temperature_data(file))
            temp_data = xr.concat(temp_data_list, dim='time')
            print(years)
            temp_data_relevant_years = temp_data.sel(time=slice("-".join([str(years[0]),"01","01"]), "-".join([str(years[1]),"12","31"])))
          
            heatwave_conditions = temp_data_relevant_years > 38+273.15 #(temp_data_relevant_years > baseline_99_percentile[model]) #& (temp_data_relevant_years > 35+273.15))
            heatwave_data = temp_data_relevant_years.where(heatwave_conditions)
            # heatwave_data = heatwave_data.where(
            # heatwave_data.rolling(time=3).count()==3)

            # Store in the dictionary
            year = file.split('_')[-1][:4]  # Extract year from filename
            heatwave_data.to_netcdf(out_file)

            heatwave_data_dict[warming_level][scenario][model] = heatwave_data

```

```python
import datetime
from pathlib import Path
import xarray as xr
import numpy as np
import os
from climada.hazard import Hazard

hazard_dir = Path("/nfs/n2o/wcr/szelie/multi_risk_data/hazards/heatwaves")
hazards_warming_level = {}
replace=True
for warming_level in ['1','2']:
    # out_file = hazard_dir / f"hazard_hw_{warming_level}_vnm.hdf5"
    # if os.path.exists(out_file) & replace==False:
    #     hazards_warming_level[warming_level] = Hazard.from_hdf5(out_file)
    haz_list = []
    for scenario in years_warming_level[warming_level]:
        for model in years_warming_level[warming_level][scenario]:

            years = years_warming_level[warming_level][scenario][model]
            if years:
                file = f'{base_path}/hw_day_{model}_{scenario}_r1i1p1_EWEMBI_{str(years[0])}0101-{years[1]}1231_vnm.nc4'
                data = xr.open_dataset(file)
                data = data.rename({'lon':'longitude', 'lat':'latitude'})


                # Count the number of non-zero days annually
                data['tas']  = xr.where(data['tas'] > 0, 1, 0)
                data = data.resample(time='Y').sum()
                data = data.fillna(0)

                haz = Hazard.from_xarray_raster(data, hazard_type='HW', intensity_unit='', intensity='tas')
                haz.event_name = [f"{str(datetime.date.fromordinal(date).year)}_{model.lower()}_{scenario}" for date in haz.date]
                haz.frequency = np.ones(len(haz.event_name))/len(haz.event_name)
                haz_list.append(haz)

        hazards_warming_level[warming_level] = Hazard.concat(haz_list)
        hazards_warming_level[warming_level].write_hdf5(hazard_dir / f"hazard_hw_{warming_level}.hdf5")
```

```python
import itertools
import numpy as np
import pycountry
import copy
from climada.util.constants import DEF_CRS
from climada.engine import Impact, ImpactCalc

from climada.util import coordinates
from climada.engine import Impact

def impact_select_country(impact, country):
    """
    Create a new impact object containing only the data for a specified country.

    This function filters an existing impact object to include only the data points
    that are within the boundaries of a given country. The function uses latitude and 
    longitude coordinates from the 'impact' object to determine the country of each 
    data point. It then filters the impact matrix and coordinates to include only those
    corresponding to the specified country.

    The country is specified by its ISO 3166-1 numeric code as a three-digit string. 
    For example, '840' for the United States, '124' for Canada, and '392' for Japan.

    Parameters:
    ----------
    impact : Impact
        The impact object to be filtered. This object should contain latitude and 
        longitude coordinates ('coord_exp') and an impact matrix ('imp_mat').

    country : str
        The three-digit string representing the ISO 3166-1 numeric country code.

    Returns:
    -------
    Impact
        A new Impact object containing only the data for the specified country.
    """
    lat, lon = np.array(impact.coord_exp).T
    countries_num = coordinates.get_country_code(lat, lon)
    countries_num = np.array([format(num, '03d') for num in countries_num])
    imp_mat = impact.imp_mat[:, countries_num == str(country)]
    coord_exp = impact.coord_exp[countries_num == str(country)]
    at_event, eai_exp, aai_agg = ImpactCalc.risk_metrics(imp_mat, impact.frequency)
    
    impact_country = Impact(
        event_id=impact.event_id,
        event_name=impact.event_name,
        date=impact.date,
        at_event=at_event,
        eai_exp=eai_exp,
        aai_agg=aai_agg,
        coord_exp=coord_exp,
        crs=DEF_CRS,
        imp_mat=imp_mat,
        frequency=impact.frequency,
        unit=impact.unit,
        frequency_unit=impact.unit,
    )
    return impact_country
```
