```python
from sklearn.utils import shuffle
import numpy as np
import matplotlib.pyplot as plt

def plot_rp_curve(impact_combined_ordered, impact_single_hazard, len_sample=500, len_total=500, n_sample=1000, max_rp=100, combinations=[('TC', 'RF')]):
    for  warming in impact_combined_ordered:
        for exposure in ['assets']:
            for combi in combinations:
                fq_list_ord = []
                fq_list_sum = []
                
                fq_list_RF = []
                fq_list_TC = []





                # Use a list comprehension to generate the values of "imp_ord" and "imp"
                # in a more concise way
                for imp_combi, imp_TC, imp_RF in [(order_events_by_indices(impact_combined_ordered[warming][exposure][combi], shuffle(np.arange(3000))[0:len_sample]),
                                     order_events_by_indices(impact_single_hazard[warming][exposure]['TC'], shuffle(np.arange(3000))[0:len_sample]),
                                     order_events_by_indices(impact_single_hazard[warming][exposure]['RF'], shuffle(np.arange(600))[0:len_sample]))
                                    for n in range(n_sample)]:

                    # Store the results of calculating the frequency curves in the appropriate lists
                    fq = imp_combi.calc_freq_curve(np.arange(1,max_rp))
                    fq_list_ord.append(fq.impact)
                    fq_TC = imp_TC.calc_freq_curve(np.arange(1,max_rp))
                    fq_list_TC.append(fq_TC.impact)
                    fq_RF = imp_RF.calc_freq_curve(np.arange(1,max_rp))
                    fq_list_RF.append(fq_RF.impact)
                    fq_list_sum.append(fq_TC.impact + fq_RF.impact)

                # Create a figure and axes using the "subplots" function from matplotlib
                fig, ax = plt.subplots(figsize=(4,3))

                # Plot the median and 5th and 95th percentile values of the "ordered" and "random" frequency curves
                ax.plot(fq.return_per, np.median(fq_list_ord, axis=0), color='purple', label= 'combined')
                ax.fill_between(fq.return_per, np.median(fq_list_ord, axis=0), np.percentile(fq_list_ord, q=95, axis=0), color='purple', alpha=0.3)
                ax.fill_between(fq.return_per, np.percentile(fq_list_ord, q=5, axis=0), np.median(fq_list_ord, axis=0), color='purple', alpha=0.3)

                ax.plot(fq.return_per, np.median(fq_list_sum, axis=0), color='purple', linestyle='dashed', label= 'Summed')
               
              # Plot the median and 5th and 95th percentile values of the "ordered" and "random" frequency curves
                ax.plot(fq.return_per, np.median(fq_list_RF, axis=0), color='blue', label= 'RF')
                ax.fill_between(fq.return_per, np.median(fq_list_RF, axis=0), np.percentile(fq_list_RF, q=95, axis=0), color='blue', alpha=0.3)
                ax.fill_between(fq.return_per, np.percentile(fq_list_RF, q=5, axis=0), np.median(fq_list_RF, axis=0), color='blue', alpha=0.3)

                # Plot the median and 5th and 95th percentile values of the "ordered" and "random" frequency curves
                ax.plot(fq.return_per, np.median(fq_list_TC, axis=0), color='red', label= 'TC')
                ax.fill_between(fq.return_per, np.median(fq_list_TC, axis=0), np.percentile(fq_list_TC, q=95, axis=0), color='red', alpha=0.3)
                ax.fill_between(fq.return_per, np.percentile(fq_list_TC, q=5, axis=0), np.median(fq_list_TC, axis=0), color='red', alpha=0.3)


```

```python
def order_events_by_indices(impact, indices):
    """
    Order event names based on given strings contained in the event names.

    ----------
    impact: Impact
        with event_name based on the given strings
    n_events: Int
        Number of events in the output. Default: 1
    list_string : list
        A list of string based on which to order the events.
        For example climate models ['miroc5','ipsl-cm5a-lr','gfdl-esm2m','hadgem2-es']
        default is None


    Raises
    ------
    AttributeError
        If no list is providing

    Returns
    -------
    impact : Impact
        Impact yearset.

    """
    impact_ordered = Impact()
    if impact_ordered.imp_mat.shape[0]>0:   
        impact_ordered.imp_mat = impact.imp_mat[indices]
    
    impact_ordered.event_name = [impact.event_name[index] for index in indices]
    impact_ordered.event_id = np.arange(len(impact_ordered.event_name))
    frequency = impact.frequency[indices]
    impact_ordered.frequency = frequency*(len(impact.event_id)/len(impact_ordered.event_id))
    impact_ordered.at_event = impact.at_event[indices]
    impact_ordered.aai_agg = np.median(impact_ordered.at_event)
    impact_ordered.coord_exp = impact.coord_exp
    impact_ordered.date = impact.date[indices]
    return impact_ordered
```
